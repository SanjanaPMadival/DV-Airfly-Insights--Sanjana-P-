```python
# List of categorical columns 
cat_cols = ['AIRLINE', 'AIRLINE_CODE', 'ORIGIN', 'ORIGIN_CITY', 
            'DEST', 'DEST_CITY', 'CANCELLATION_CODE']

# Loop through each column and convert it to 'category' dtype
for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
# Downcast floats to float32
float_cols = df.select_dtypes(include=['float64']).columns
df[float_cols] = df[float_cols].astype('float32')
```

```python
# Downcast ints to smaller type if possible
int_cols = df.select_dtypes(include=['int64']).columns
df[int_cols] = df[int_cols].apply(pd.to_numeric, downcast='unsigned')
```

```python
# Drop unnecessary columns
df = df.drop(columns=['AIRLINE_DOT', 'DOT_CODE', 'FL_NUMBER'])
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# --- Convert to Appropriate Types ---
categorical_cols = ['AIRLINE', 'AIRLINE_CODE', 'ORIGIN', 'ORIGIN_CITY',
                    'DEST', 'DEST_CITY', 'CANCELLATION_CODE']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
# ---Convert numeric columns to smaller float types for memory efficiency
numeric_cols = ['DEP_DELAY', 'ARR_DELAY', 'TAXI_OUT', 'TAXI_IN', 
                'CRS_ELAPSED_TIME', 'ELAPSED_TIME', 'AIR_TIME', 'DISTANCE',
                'DELAY_DUE_CARRIER','DELAY_DUE_WEATHER','DELAY_DUE_NAS',
                'DELAY_DUE_SECURITY','DELAY_DUE_LATE_AIRCRAFT']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], downcast='float')
```

```python
# --- Datetime Conversion ---
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], errors='coerce')
```

```python
# Convert to datetime, then extract only the time (HH:MM, no seconds)
df['dep_time'] = pd.to_datetime(
    df['DEP_TIME'].fillna(0).astype(int).astype(str).str.zfill(4),
    format='%H%M',
    errors='coerce'
).dt.time

df['arr_time'] = pd.to_datetime(
    df['ARR_TIME'].fillna(0).astype(int).astype(str).str.zfill(4),
    format='%H%M',
    errors='coerce'
).dt.time

```

```python
# --- Temporal Features ---
df['Month'] = df['FL_DATE'].dt.month
df['DayOfWeek'] = df['FL_DATE'].dt.dayofweek.replace({0:7}) # 7=Monday
df['Quarter'] = df['FL_DATE'].dt.quarter
df['IsWeekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
df['YEAR'] = df['FL_DATE'].dt.year
print(df['YEAR'].unique())
print(df['YEAR'].value_counts().sort_index())
```

```python
# --- Route Feature ---
df['Route'] = df['ORIGIN'].astype(str) + '-' + df['DEST'].astype(str)
```

```python
# --- Distance Binning ---
bins = [0, 500, 1500, 3000, np.inf]
labels = ['Short', 'Medium', 'Long', 'Very_Long']
df['Distance_Bin'] = pd.cut(df['DISTANCE'], bins=bins, labels=labels)
```

```python
# --- Handle Nulls & Outliers ---
delay_cols = ['DEP_DELAY', 'ARR_DELAY', 'DELAY_DUE_CARRIER', 
              'DELAY_DUE_WEATHER','DELAY_DUE_NAS','DELAY_DUE_SECURITY',
              'DELAY_DUE_LATE_AIRCRAFT']
for col in delay_cols:
    df[col] = df[col].fillna(0)
    # Optional: Cap extreme outliers
    df[col] = df[col].clip(upper=df[col].quantile(0.99))
```

```python
# Cancellation Handling
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].cat.add_categories('Unknown')
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].fillna('Unknown')
```

```python
# Binary flags for cancellation reasons
df['Cancelled_Carrier'] = np.where(df['CANCELLATION_CODE']=='A', 1, 0)
df['Cancelled_Weather'] = np.where(df['CANCELLATION_CODE']=='B', 1, 0)
df['Cancelled_NAS'] = np.where(df['CANCELLATION_CODE']=='C', 1, 0)
df['Cancelled_Security'] = np.where(df['CANCELLATION_CODE']=='D', 1, 0)
```

```python
# --- Airport Congestion Feature ---
airport_flights = df.groupby(['ORIGIN', 'FL_DATE']).size().reset_index(name='Flights_Per_Origin')
df = df.merge(airport_flights, on=['ORIGIN','FL_DATE'], how='left')
```

```python
# --- Rolling Metrics Example ---
df['Rolling7D_DepDelay'] = df.groupby('AIRLINE')['DEP_DELAY'].transform(lambda x: x.rolling(7, min_periods=1).mean())
```

```python
# Function to return the main delay cause and its minutes for each row
def get_delay_reason_and_time(row):
    if row['DELAY_DUE_CARRIER'] > 0:
        return "CARRIER", int(row['DELAY_DUE_CARRIER'])
    elif row['DELAY_DUE_WEATHER'] > 0:
        return "WEATHER", int(row['DELAY_DUE_WEATHER'])
    elif row['DELAY_DUE_NAS'] > 0:
        return "NAS", int(row['DELAY_DUE_NAS'])
    elif row['DELAY_DUE_SECURITY'] > 0:
        return "SECURITY", int(row['DELAY_DUE_SECURITY'])
    elif row['DELAY_DUE_LATE_AIRCRAFT'] > 0:
        return "LATE_AIRCRAFT", int(row['DELAY_DUE_LATE_AIRCRAFT'])
    else:
        return "NO_DELAY", 0

# Apply the function row-wise and create two new columns: DELAY_CAUSE & DELAY_MINUTES
df[['DELAY_CAUSE', 'DELAY_MINUTES']] = df.apply(get_delay_reason_and_time, axis=1, result_type='expand')
```

```python
delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

# Drop in place
df.drop(columns=delay_cols, inplace=True)
```

```python
# -- Columns to drop -- 
drop_cols = [
    "AIRLINE_CODE",
    "ORIGIN_CITY", "DEST_CITY",
    "CRS_DEP_TIME", "CRS_ARR_TIME", "DEP_TIME", "ARR_TIME",
    "CRS_ELAPSED_TIME", "ELAPSED_TIME", "AIR_TIME",
    "TAXI_OUT", "TAXI_IN", "WHEELS_OFF", "WHEELS_ON",
    "CANCELLATION_CODE"
]

# Drop safely (ignore if column missing)
df = df.drop(columns=[col for col in drop_cols if col in df.columns])

print("Remaining columns:", df.columns.tolist())
print("Shape after dropping:", df.shape)
```
